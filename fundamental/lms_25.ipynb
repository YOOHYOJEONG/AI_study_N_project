```python
from multiprocessing import Pool
from os import getpid

def double(i):
    print("I'm processing ", getpid())    # pool ì•ˆì—ì„œ ì´ ë©”ì†Œë“œê°€ ì‹¤í–‰ë  ë•Œ pidë¥¼ í™•ì¸í•´ ë´…ì‹œë‹¤.
    return i * 2

with Pool() as pool:
      result = pool.map(double, [1, 2, 3, 4, 5])
      print(result)
```

# ë³‘ë ¬ í”„ë¡œê·¸ë˜ë° ì—°ìŠµ
- concurrent.futures ëª¨ë“ˆì˜ ProcessPoolExecutorë¥¼ ì´ìš©í•´ì„œ ë©€í‹°í”„ë¡œì„¸ìŠ¤ë¥¼ êµ¬í˜„ì„ ì—°ìŠµ.
- Executor ê°ì²´ì˜ map() í•¨ìˆ˜
- ProcessPoolExecutor ë¶€ë¶„

```python
import math
import concurrent
import time 


PRIMES = [
    112272535095293,
    112582705942171,
    112272535095293,
    115280095190773,
    115797848077099,
    1099726899285419]

#ì†Œìˆ˜(prime) íŒë³„ ë¬¸ì œë¡œ PRIMES ë³€ìˆ˜ì— ì„ ì–¸ëœ ìˆ«ìë“¤ì´ ì†Œìˆ˜ì¸ì§€ ì•„ë‹Œì§€ë¥¼ íŒë³„í•˜ëŠ” í•¨ìˆ˜.
def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False

    sqrt_n = int(math.floor(math.sqrt(n)))
    for i in range(3, sqrt_n + 1, 2):
        if n % i == 0:
            return False
    return True


#main() í•¨ìˆ˜ë¥¼ ë”°ë¡œ ì‘ì„±í•˜ì—¬ ì†Œìˆ˜ íŒë³„ í•¨ìˆ˜(is_prime)ë¥¼ í˜¸ì¶œ
#ë³‘ë ¬ì²˜ë¦¬ì™€ ë‹¨ì¼ì²˜ë¦¬ ë¹„êµ ìœ„í•´ main í•¨ìˆ˜ë¥¼ ì•„ë˜ì™€ ê°™ì´ ìˆ˜ì •.
#í”„ë¡œíŒŒì¼ë§ì„ ìœ„í•œ ì‹œê°„ ê³„ì‚° ì½”ë“œë¥¼ ì¶”ê°€
#ë‹¨ì¼ì²˜ë¦¬ë¡œ ìˆ˜í–‰í–ˆì„ ë•Œì˜ ì½”ë“œë¥¼ ì¶”ê°€
#ë‹¨ì¼ì²˜ë¦¬ í”„ë¡œíŒŒì¼ë§ì„ ìœ„í•œ ì‹œê°„ ê³„ì‚° ì½”ë“œë¥¼ ì¶”ê°€
def main():
    print("ë³‘ë ¬ì²˜ë¦¬ ì‹œì‘")
    start = time.time()
   
    #concurrent.futures ë¼ì´ë¸ŒëŸ¬ë¦¬ì˜ í”„ë¡œì„¸ìŠ¤ í’€ì—ì„œ ë™ì‘í•˜ê²Œ í•˜ê¸° ìœ„í•´ with ë¬¸ ì‚¬ìš©.
    with concurrent.futures.ProcessPoolExecutor() as executor:
        
        #map() í•¨ìˆ˜ë¥¼ ProcessPoolExecutor() ì¸ìŠ¤í„´ìŠ¤ì—ì„œ ìƒì„±ëœ executor ì—ì„œ ì‹¤í–‰.
        for number, prime in zip(PRIMES, executor.map(is_prime, PRIMES)):
            print('%d is prime: %s' % (number, prime))
        end = time.time()
        print("ë³‘ë ¬ì²˜ë¦¬ ìˆ˜í–‰ ì‹œê°", end-start, 's')

        start = time.time()
        for number, prime in zip(PRIMES, map(is_prime, PRIMES)):
            print('%d is prime: %s' % (number, prime))
        end = time.time()
        print("ë‹¨ì¼ì²˜ë¦¬ ìˆ˜í–‰ ì‹œê°", end-start, 's')

print(" â£\nğŸŒ²ğŸ¦•.......")
        
        
if __name__ == '__main__':
    main()
```
